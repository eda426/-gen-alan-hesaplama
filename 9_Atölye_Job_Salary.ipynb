.info() metodu, genellikle pandas kütüphanesi ile kullanılan ve bir DataFrame hakkında özet bilgi sağlayan bir fonksiyondur. Bu özet bilgi, veri setinin genel yapısını anlamak için oldukça faydalıdır.

    .info() Metodunun Sağladığı Bilgiler:

- Veri Tipleri: Hangi sütunun hangi veri tipine (örneğin, int, float, object (string), datetime) sahip olduğunu gösterir.
- Sütun İsimleri: Veri setindeki tüm sütunların isimlerini listeler.
- Null (Eksik) Değerler: Her sütundaki eksik (null) değerlerin sayısını gösterir.
- Sütun Sayısı ve Satır Sayısı: DataFrame'deki toplam satır ve sütun sayısını belirtir.
- Bellek Kullanımı: DataFrame'in bellekte ne kadar yer kapladığını gösterir.

```python
df.info()
```

.describe() metodu, genellikle pandas kütüphanesi ile kullanılan ve bir DataFrame'in sayısal sütunları hakkında özet istatistiksel bilgi sağlayan bir fonksiyondur. Bu metod, veri setinin istatistiksel özelliklerini hızlı bir şekilde anlamak için oldukça kullanışlıdır.


    .describe() Metodunun Sağladığı Bilgiler:
- count: Her sütundaki toplam veri noktası sayısını gösterir.
- mean: Her sütunun ortalama değerini hesaplar.
- std: Her sütunun standart sapmasını gösterir.
- min: Her sütundaki en küçük değeri belirtir.
- 25% (1. çeyrek): Her sütundaki değerlerin %25'inin altında kaldığı değeri gösterir.
- 50% (Medyan): Her sütundaki değerlerin %50'sinin altında kaldığı değeri belirtir (medyan).
- 75% (3. çeyrek): Her sütundaki değerlerin %75'inin altında kaldığı değeri gösterir.
- max: Her sütundaki en büyük değeri belirtir.

**matplotlib.pyplot**

   **Çubuk Grafikler için Bilinmesi Gerekenler**
- Kategorik Veri Temsili: Çubuk grafikler, kategorik verileri (örneğin, farklı şehirler, ürün türleri) net bir şekilde görselleştirmek için kullanılır. Her çubuk, belirli bir kategoriyi temsil eder.
- Değer Karşılaştırması: Çubukların uzunluğu veya yüksekliği, kategoriler arasındaki değerleri doğrudan karşılaştırmak için görsel bir yol sağlar. Uzun çubuklar yüksek değerleri, kısa çubuklar düşük değerleri temsil eder.
- Netlik ve Anlaşılabilirlik: Çubuk grafikler, verileri açık ve anlaşılır bir şekilde sunar. Kategoriler arasındaki farklar kolayca gözlemlenebilir, bu da verilerin hızlıca analiz edilmesini sağlar.
- Renk ve Etiketler: Çubuk grafiklerde çubuklar farklı renklerde olabilir ve her çubuğun üzerinde veya altında etiketler bulunur. Bu, verilerin daha iyi anlaşılmasını ve grafik üzerinde kolayca tanımlanmasını sağlar.

- plt.figure(): Yeni bir grafik alanı (canvas) oluşturur.
- figsize=(x, y): Grafik alanının genişlik ve yüksekliğini (inch cinsinden) belirler.
- plt.bar(): Dikey çubuklardan oluşan bar (çubuk) grafiği çizer.
- x=: X eksenine hangi sütunun geleceğini belirtir (kategoriler, tarih vb.).
- height=: Her çubuğun yüksekliğini belirleyen sütunu seçer (sayısal değerler).
- data=: Grafiğe temel olacak veri kaynağı DataFrame’i belirtir.
- plt.xticks(rotation=…): X eksenindeki etiketlerin dönme açısını ayarlar (örneğin 90 = dikey yazı).

```python
#grafiğin veri kaynağı burada
group_category_salary = df.groupby("job_category")[["salary_in_usd"]].mean().sort_values(by="salary_in_usd",ascending=False).reset_index()
group_category_salary
```

```python
#matplotlib ile çubuk grafiği oluşturalım

plt.figure(figsize=(15,5)) #Grafiğini boyutunu belirler
plt.bar(x = "job_category", height = "salary_in_usd", data = group_category_salary)
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(15,5)) #Grafiğini boyutunu belirler
sns.barplot(x = "job_category", y = "salary_in_usd", data = group_category_salary)
plt.xticks(rotation=90);
```

Boxplot (Kutu Grafiği) Nedir?
Boxplot, bir veri setinin dağılımını özetleyen istatistiksel bir grafiktir.

*Medyan ve Çeyrekler:*
- Kutunun ortasındaki çizgi medyanı, kenarları 1. (Q1) ve 3. (Q3) çeyrekleri gösterir.

*IQR (Çeyrekler Arası Mesafe):*
- Kutunun boyu Q1 ile Q3 arasındaki farktır; verinin ortadaki %50’sini temsil eder.

*Aykırı Değerler:*
- IQR’ın 1.5 katı dışındaki değerler aykırı kabul edilir ve noktalarla gösterilir.

*Whiskers (Sorguçlar):*
- Aykırı olmayan minimum ve maksimum değerleri gösteren çizgilerdir.

*Simetriklik:*
- Medyan ortadaysa ve sorguçlar eşitse dağılım simetriktir; değilse asimetridir.

agg() metodunu, bir veri çerçevesinde veya serisinde çeşitli toplu fonksiyonlar uygulamak için kullanabilirsiniz. Bu fonksiyonlar, ortalama (mean), toplam (sum), medyan (median), minimum (min), maksimum (max) gibi işlemler olabilir.

```python
df.groupby("job_category")[["salary_in_usd"]].agg(["mean","max","min"]).reset_index()
```
