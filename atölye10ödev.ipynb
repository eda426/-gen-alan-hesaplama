```python
#Seçtiğiniz bir veri seti üzerinde veri analizi yapın
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
 #dosyayı okuyalım
df=pd.read_csv("netflix_titles.csv")
df.head()
```

```python
#Film ve dizi sayısını görmek için
df["type"].value_counts()
```

```python
#Yıllara Göre Film ve Dizi Sayısı
sns.countplot(x="release_year",hue="type",data=df)
plt.title('Yıllara Göre Film ve Dizi Sayısı')
plt.xticks(rotation=45)
```

```python
#En fazla içerik olan ilk 10 ülke:
df["country"].value_counts().head(10)
```

```python
#film-dizi sayısına histogram grafiği ile bakalım
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df["release_year"])
plt.title("Film-Dizi Sayısına Histogram Grafiği")
```
